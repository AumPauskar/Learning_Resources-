```python
tot_resp=df["Country"].value_counts()
know_py=grp['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
py_per=pd.concat([tot_resp,know_py],axis='columns',sort=False)
py_per
```

```python
py_per.rename(columns={'Country':'TotalRespondents','LanguageWorkedWith':'KnowPython'},inplace=True)
py_per
```

```python
py_per['PercentageKnowPython']=(py_per['KnowPython']*100/py_per['TotalRespondents'])
```
